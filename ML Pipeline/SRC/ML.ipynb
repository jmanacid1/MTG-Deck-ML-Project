```python
import pandas as pd
import sklearn as sk
```

```python
file_path = "../../Data Processing/Outputs/processed_fdn_draft_data"
df = pd.read_pickle(file_path)
```

```python
df.head()
```

```python
maindeck_df = df.loc[df['pick_maindeck_rate'] == 1]
agg_maindeck_df = maindeck_df.groupby('draft_id')['pick'].apply(list).reset_index(name = 'maindeck')
agg_maindeck_df.head()
```

```python
corpus = agg_maindeck_df['maindeck']
print(corpus)

vectorizer = sk.TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
vectorizer.get_feature_names_out()
```
